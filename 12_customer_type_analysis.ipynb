```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    r"C:\Users\Hp\Desktop\Travel-Tourism-Customer-Retention-Analysis\data\processed\hotel_bookings_validated.csv"
)

df.shape
```

```python
df['customer_type'].value_counts()
```

## Customer Type Distribution

Customer types represent different booking behaviors and customer relationships with the hotel.

Understanding customer composition helps identify which customer groups contribute most to bookings.

```python
(df['customer_type'].value_counts(normalize=True) * 100).round(2)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='customer_type'
)

plt.title("Customer Type Distribution")
plt.xlabel("Customer Type")
plt.ylabel("Number of Bookings")

plt.show()
```

## Customer Type Overview

The chart shows the distribution of booking records across different customer categories.

```python
customer_cancel_rate = (
    df.groupby('customer_type')['is_canceled']
      .mean()
      .mul(100)
      .round(2)
)

customer_cancel_rate
```

```python
customer_cancel_rate.plot(
    kind='bar'
)

plt.title("Cancellation Rate by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Cancellation Rate (%)")

plt.show()
```

## Cancellation Analysis

Cancellation rates are compared across customer categories to identify which customer segments are more likely to cancel bookings.

## Findings

- Customer type distribution was analyzed.
- Cancellation rates were calculated for each customer type.
- Customer segments show different cancellation behavior.
- Some customer categories demonstrate higher cancellation tendencies than others.

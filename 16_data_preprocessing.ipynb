## Data Preprocessing

Machine learning models require numerical input.

Categorical variables are converted into numeric values using Label Encoding.

```python
categorical_columns = df.select_dtypes(include='object').columns

categorical_columns
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
for column in categorical_columns:
    df[column] = encoder.fit_transform(df[column])

df.head()
```

## Encoding Summary

All categorical variables have been converted into numerical values.

The dataset is now compatible with machine learning algorithms.

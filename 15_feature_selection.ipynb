## Dataset Overview

The validated hotel booking dataset is loaded for machine learning preparation.

The objective is to identify the most relevant features that influence booking cancellation.

```python
target = 'is_canceled'

print("Target Variable:", target)
```

```python
selected_features = [
    'hotel',
    'lead_time',
    'arrival_date_month',
    'stays_in_weekend_nights',
    'stays_in_week_nights',
    'adults',
    'children',
    'babies',
    'meal',
    'country',
    'market_segment',
    'distribution_channel',
    'is_repeated_guest',
    'previous_cancellations',
    'previous_bookings_not_canceled',
    'reserved_room_type',
    'assigned_room_type',
    'booking_changes',
    'deposit_type',
    'customer_type',
    'adr',
    'required_car_parking_spaces',
    'total_of_special_requests',
    'total_stay_nights',
    'total_guests',
    'booking_changed',
    'has_previous_cancellation',
    'special_request_flag'
]

len(selected_features)
```

```python
X = df[selected_features]

X.head()
```

```python
y = df[target]

y.head()
```

```python
print("Feature Matrix:", X.shape)
print("Target Variable:", y.shape)
```

```python
print("Missing Values in Features:")
print(X.isnull().sum().sum())

print("Missing Values in Target:")
print(y.isnull().sum())
```

## Feature Selection

Relevant predictor variables were selected based on previous exploratory data analysis and domain knowledge.

The target variable for prediction is **is_canceled**.

```python
selected_df = pd.concat([X, y], axis=1)

selected_df.to_csv(
    r"C:\Users\Hp\Desktop\Travel-Tourism-Customer-Retention-Analysis\data\processed\hotel_bookings_selected_features.csv",
    index=False
)

selected_df.head()
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

## Target Variable

The target variable for this project is:

is_canceled

0 = Booking Not Cancelled

1 = Booking Cancelled

```python
df['is_canceled'].value_counts()
```

```python
cancellation_percent = (
    df['is_canceled']
    .value_counts(normalize=True)
    * 100
)

cancellation_percent
```

## Cancellation Distribution

This analysis helps understand the proportion of cancelled and retained bookings.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='is_canceled',
    data=df
)

plt.title("Booking Cancellation Distribution")

plt.xlabel("Cancellation Status")

plt.ylabel("Count")

plt.show()
```

```python
cancel_counts = df['is_canceled'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    cancel_counts,
    labels=[
        "Not Cancelled",
        "Cancelled"
    ],
    autopct="%1.1f%%"
)

plt.title("Cancellation Percentage")

plt.show()
```

```python
cancelled = (
    df['is_canceled']
    .value_counts()[1]
)

not_cancelled = (
    df['is_canceled']
    .value_counts()[0]
)

print("Cancelled Bookings:", cancelled)

print("Not Cancelled Bookings:", not_cancelled)
```

```python
cancellation_rate = (
    df['is_canceled']
    .mean()
    * 100
)

print(
    "Cancellation Rate:",
    round(cancellation_rate,2),
    "%"
)
```

```python
retention_rate = (
    (1 - df['is_canceled'].mean())
    * 100
)

print(
    "Retention Rate:",
    round(retention_rate,2),
    "%"
)
```

## Business Interpretation

The cancellation rate indicates the percentage of bookings lost before customer arrival.

A higher cancellation rate directly impacts hotel revenue and occupancy planning.

Understanding cancellation behavior is essential for customer retention strategies and dynamic pricing decisions.

```python
df['booking_changes'].describe()
```

## Booking Changes Overview

Booking changes represent modifications made by customers after the initial reservation.

Understanding booking modifications helps identify customer commitment and cancellation behavior.

```python
booking_change_cancel = (
    df.groupby('booking_changed')['is_canceled']
      .mean()
      .mul(100)
      .round(2)
)

booking_change_cancel
```

```python
booking_change_cancel.plot(
    kind='bar'
)

plt.title("Cancellation Rate by Booking Change Status")
plt.xlabel("Booking Changed")
plt.ylabel("Cancellation Rate (%)")

plt.show()
```

## Booking Modification Analysis

Cancellation rates are compared between modified and non-modified bookings.

## Special Requests Overview

Special requests reflect additional customer requirements during booking.

These requests may indicate customer engagement and commitment.

```python
special_request_cancel = (
    df.groupby('special_request_flag')['is_canceled']
      .mean()
      .mul(100)
      .round(2)
)

special_request_cancel
```

```python
special_request_cancel.plot(
    kind='bar'
)

plt.title("Cancellation Rate by Special Request Status")
plt.xlabel("Special Request")
plt.ylabel("Cancellation Rate (%)")

plt.show()
```

## Findings

- Booking modifications were analyzed against cancellation behavior.
- Special requests were examined as indicators of customer engagement.
- Customers with booking changes may demonstrate different cancellation patterns.
- Special requests appear to influence booking retention behavior.

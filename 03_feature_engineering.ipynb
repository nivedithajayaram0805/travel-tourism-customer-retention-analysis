```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    r"C:\Users\Hp\Desktop\Travel-Tourism-Customer-Retention-Analysis\data\cleaned\hotel_bookings_cleaned.csv"
)

df.head()
```

```python
df.shape
```

```python
df['total_stay_nights'] = (
    df['stays_in_weekend_nights']
    + df['stays_in_week_nights']
)

df[['stays_in_weekend_nights',
    'stays_in_week_nights',
    'total_stay_nights']].head()
```

```python
df['total_stay_nights'].describe()
```

```python
df['total_guests'] = (
    df['adults']
    + df['children']
    + df['babies']
)

df[['adults','children','babies','total_guests']].head()
```

```python
df['total_guests'].describe()
```

```python
df['booking_changed'] = np.where(
    df['booking_changes'] > 0,
    1,
    0
)

df['booking_changed'].value_counts()
```

```python
df['has_previous_cancellation'] = np.where(
    df['previous_cancellations'] > 0,
    1,
    0
)

df['has_previous_cancellation'].value_counts()
```

```python
df['special_request_flag'] = np.where(
    df['total_of_special_requests'] > 0,
    1,
    0
)

df['special_request_flag'].value_counts()
```

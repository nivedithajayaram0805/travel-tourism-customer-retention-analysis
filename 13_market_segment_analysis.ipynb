```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    r"C:\Users\Hp\Desktop\Travel-Tourism-Customer-Retention-Analysis\data\processed\hotel_bookings_validated.csv"
)

df.shape
```

```python
df['market_segment'].value_counts()
```

## Market Segment Distribution

Market segments indicate the source through which customers book hotel rooms.

Understanding booking sources helps hotels optimize marketing strategies and customer acquisition efforts.

```python
(df['market_segment'].value_counts(normalize=True) * 100).round(2)
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x='market_segment',
    order=df['market_segment'].value_counts().index
)

plt.title("Market Segment Distribution")
plt.xlabel("Market Segment")
plt.ylabel("Number of Bookings")

plt.xticks(rotation=45)

plt.show()
```

## Booking Source Overview

The chart displays how bookings are distributed across different market segments.

```python
segment_cancel_rate = (
    df.groupby('market_segment')['is_canceled']
      .mean()
      .mul(100)
      .round(2)
      .sort_values(ascending=False)
)

segment_cancel_rate
```

```python
segment_cancel_rate.plot(
    kind='bar'
)

plt.title("Cancellation Rate by Market Segment")
plt.xlabel("Market Segment")
plt.ylabel("Cancellation Rate (%)")

plt.show()
```

## Cancellation Analysis

Cancellation rates are compared across market segments to identify which booking sources contribute most to cancellations.

## Findings

- Market segment distribution was analyzed.
- Cancellation rates vary significantly across booking sources.
- Some segments contribute more cancellations than others.
- Customer acquisition channels influence cancellation behavior.
